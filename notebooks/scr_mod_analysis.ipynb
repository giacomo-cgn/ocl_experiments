```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# read file line by line ../results/scr_split_cifar100_20_2000/0/logs_continual.json
with open('../results/scr_split_cifar100_20_2000/0/logs_continual.json') as f:
    lines = f.readlines()
    dict_list = [json.loads(line) for line in lines]
```

```python
df = pd.DataFrame(dict_list)
df.info()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(15, 6))

for i in range(20):
    idx = str(i)
    # add 0 char in front if str length is =1
    if len(str(i)) == 1:
        idx = '0' + str(i)
    else:
        idx = str(i)
    plt.plot(df['Top1_Acc_Exp/eval_phase/valid_stream/Task000/Exp0' + idx], label='exp {}'.format(idx))

plt.show()
```
